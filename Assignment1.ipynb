i. Find the mean $m$ and the standard deviation $\sigma$ of this population.
+ *You can use the methods numpy.ndarray.mean() and numpy.ndarray.std() to calculate the mean and standard deviation.*

```python
alien=[2,3,6,8,11,18]
print ("This is the mean: ",np.mean(alien))
print ("This is the standard deviation: ", np.std(alien))
```

> ii. List all possible samples of two aliens without replacement, and find each mean. Do the same with samples of four aliens.
+ *You can use the method itertools.combinations(v,n) to find all possible samples of a vector v taking n elements at a time.*

```python
from itertools import combinations

##We do the exercise for samples of 2 aliens

ls = (list(combinations(alien,2)))
print ("this are all the possible combinations of 2 values without replacement:",ls)
means2=[]
for i in ls :
    means2.append(np.mean(i))

print ("this are the means:" , means2)
```

```python


ls = (list(combinations(alien,4)))
print ("this are all the possible combinations of 4 values without replacement:",ls)
means4=[]
for i in ls :
    means4.append(np.mean(i))

print ("this are the means:" , means4)
```
