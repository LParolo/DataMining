## 4.1 *k*-means clustering
In this part of the exercise we will investigate *k*-means clustering. In *k*-means each of the data points are assigned to the cluster in closest proximity according to some measure of distance between cluster centers and data points. When the distance is given by the squared euclidian distance, the centers are also called centroids. Once the data points have been assigned, each cluster center is updated to be placed at the center of the data points that are assigned to the cluster. This continues iteratively, usually until the assignment of data points to centers no longer changes or until a maximal number of iterations is reached.

Cluster validity measure are usefull to determine the best n for K=n (with K=the number of clusters),  as they define mathematically how well the clustering performs in relations to K . As we see in the plots above, this is totally dependent by the type of data we have, and how the clustering interprets the dataset.

For K=4 , both the entropy and the rand start to level off,while this doesn't happen for the purity and the jaccard. In fact the purity increase slightly for K>4 (except for the fisrt dataset,which levels off at k=4) , while the jaccard seems to decrease for every dataset for k>4. In general , the clustering seems to work very well for the first dataset, while has worst results for the other 3 datasets. 

Unluckily (?) the choice of K remains a subjective preference, as there is no measures that predicts exactly the best K. We can only use these methods to have a glimpse of what could work better, but we still need the human imagination and especially our ability to understand the meaning of the clusters and the dataset, beyond its mathamtical representation. 
Therefore we will have to find an adequate compromise between complexity (or simplicity) of our model and its ability to properly describe the data.
